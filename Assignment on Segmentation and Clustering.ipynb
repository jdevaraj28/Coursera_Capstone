```python
import requests # library to handle requests
import pandas as pd # library for data analsysis
import numpy as np # library to handle data in a vectorized manner
```

Import beautifulsoup from bs4

```python
from bs4 import BeautifulSoup
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
req = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
soup=BeautifulSoup(req.text,'lxml')
```

```python
print(soup.title)
```

```python
tab = str(soup.table)
```

```python
tab
```

```python
from IPython.display import display_html
```

```python
display_html(tab,raw=True)
```

```python
dfs = pd.read_html(tab)
df=dfs[0]
df.head()

```

```python
# Data preprocessing and cleansing
df1 = df[df.Borough != 'Not assigned']
df1
```

Combining the neighbourhoods with same Postalcode

```python
df1.rename(columns={'Postal Code':'Postcode'},inplace=True)
```

After renaming, combine the neighbourhood with same postalcode


```python
df2 = df1.groupby(['Postcode','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

```

```python
#if Neighbourhood is not assigned, replace by borough name
df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned',df2['Borough'], df2['Neighbourhood'])

df2
```

print the number of rows of your dataframe

```python
df1.shape
```

Importing the csv file conatining the latitudes and longitudes for various neighbourhoods in Canada


```python
lat_lon_toronto = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon_toronto.head()
```

Merging the two tables for getting the Latitudes and Longitudes for various neighbourhoods in Canada


```python
lat_lon_toronto.rename(columns={'Postal Code':'Postcode'},inplace=True)
df3 = pd.merge(df2,lat_lon_toronto,on='Postcode')
df3.head()
```

# Cluster Neighborhoods

```python
kcluster=5
toronto_clustering = df3.drop(['Postcode','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = kcluster,random_state=0).fit(toronto_clustering)
kmeans.labels_
df3.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
df3
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df3['Latitude'], df3['Longitude'], df3['Neighbourhood'], df3['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
